## Imports

```python
import pandas as pd
```

## Read in Data

```python
coef_df = pd.read_csv('lme_coefficients.csv')
gun_topic_scores = pd.read_csv('gun_law_topic_scores.csv')
```

```python
coef_df.head()
```

```python
gun_topic_scores.head()
```

## Computing the Effect of Each Gun Law

```python
law_effect = coef_df.merge(gun_topic_scores, on='state')
```

```python
topic_col_names = [f'topic_{i}' for i in range(20)]
coef_col_names = [f'coef_topic_{i}' for i in range(20)]
```

```python
law_effect['effect'] = law_effect[topic_col_names].mul(
    law_effect[coef_col_names].values, axis=0).sum(axis=1)
```

```python
law_effect.head()
```

```python
law_effect.to_csv('gun_law_topics_effects.csv', index=False)
```
